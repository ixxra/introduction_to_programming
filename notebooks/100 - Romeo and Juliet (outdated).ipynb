Let's split the text into words, and next see a few of the words obtained:

```python
words = text.split()
print words[:30]
print len(words), 'total words in the play'
```

There are 25640 words in the play, however, our word list if far from optimum for data analysis, since there are words like **'scene,'** (did you notice the trailing comma?). Moreover, there are words like **'In'** and **'in'** that would be counted as different words since in general, in computer science, uppercase and lowercase words are considered differents:

```python
print words.count('in'), 'times "in"'
print words.count('In'), 'times "In"'
```

To remedy this we must procceed as follows: first we need to make all the text lowercase (or uppercase) and then we need to split wisely:

```python
text = open('data/romeo_juliet').read().lower()
words = re.split('\W', text)
print words[:30]
```

Did you notice the command to make the text *lower case*? (*hint:* Read carefully the first line). In the second line, we invoked the package *re* imported previously. In this case, *re.split* is a command in the package that splits the text with help of a [regular expression](http://en.wikipedia.org/wiki/Regular_expression), which is the *'\W'*. This regular expression means *split the text wherever you find a non word character*

Note that in the resulting list, there are many *empty words*, recognized by the *''* symbol. We have to filter those empty strings. In order to do so, we will use what is called a [list comprehension](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions)

Be sure to [read](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions) the documentation on list comprehensions until you understand what it is about.

```python
words = [word for word in words if len(word) > 0]
print words[:30]
print len(words), 'total words in the play'
```

Let's count occurrences of *different* words. In python, we can count a list entries without repetition by making a *set*:

```python
different_words = set(words)
print len(different_words), 'different words in the play'
```

From the 25640 words in Romeo and Juliet, there are only 3538 non repeated words, that is roughly the 14% of the play. 

**A word of caution:** 
    
    If you calculate the proportion in python, be sure to write something like
    
    3583 / 25640.
    
    with the decimal point, otherwise, you would obtain the awkward result of 0. This is due to the fact that python converts "integer division" into an integer result. 

[Read](https://docs.python.org/2/tutorial/introduction.html#numbers) the tutorial until you understand this.

Let's count the number of times each word is repeated in the python style.

```python
frequencies = [words.count(word) for word in different_words]
print zip(different_words, frequencies)[:30]
```

The *zip* function takes two or more lists and *zip'em* in a cartesian product. So, from the fisrt thirty non repeated words, *hath* is the most repeated, appearing sixty three times in the play (can you find it?)

Let's order the words, in decreasing order of repeatition, so we can gain (qualitatively) insight about our data:

```python
table = zip(different_words, frequencies)
table.sort(key=lambda (x,y): y, reverse=True)

print table[:30]
```

Those are the first 30 most repeated words in the play, *and* being the most popular, and *love* being the less. The command we used to make the ordering was *sort*, the *key* argument is what is called a [lambda expression](https://docs.python.org/2/tutorial/controlflow.html#lambda-expressions), which is a condensed way to say that, on each pair *(x, y)*, in the zip, the second entry should be used to make the ordering. 

(**Question:** Can you tell what is the *reverse* keyword for?)

If we want to be more explicit in the representation of the data, we could use a *cicle* to print the data with more fashion:

```python
print '{w:>12}    {f:>8}    {h}'.format(w="Words", f="Frecuencies", h="Histogram")

for word, frequency in table:
    stars = '*' * (frequency / 10)
    print '{w:>12}       {f:>03}         {h}'.format(w=word, f=frequency, h=stars)
```

That's a pretty cool text histogram!. What we are doing in the previous entry goes as follows:

First, we print a header for the histogram with the [format](https://docs.python.org/2/library/string.html#format-examples) function. The syntax 

    '{w:>12}    {f:>8}    {h}'
        
means that there are three data to be substituted in the string, with the placeholders *w*, *f* and *h*. 

The symbol *{w:>12}* means that the first string (the *word*) must have reserved twelve spaces in order to be printed, and that it should be *right justified* (due to the *>* in the format specification). I hope you grab the idea here. Nevertheless, you should read the documentation, in case you didn't get it.

After printing the header, we go into a *cycle*, and, just as the code says, *for each* pair of *word* and *frequency* in the *table*, we print another formated string, very simmilar to the header.

Note that we defined a *stars* variable by multiplying a string with a *lone star*, `*` times an integer number, obtained by dividing the frequency between 10. 

The net result of the previous operations is: first, we rescale the frequency to one order of magnitude less, that is, the most popular word won't be repeated 718 times, but *718 / 10 = 71* times (you remember integer division, right?), and then we multiply this number times a *star*, which in python means to repeat the given text as many times as the multiplication indicates.

*Clever, isn't it?* ;)

This notebook is going to lengthy. In the next section we will just *numpy* and *matplotlib* to improve our histogram, but, before we should save our fresh data. 

We are going to save the frequency table in a format called [csv](http://en.wikipedia.org/wiki/Comma-separated_values) (comma separated values). It is assumed that you are running the code in this notebook with *ipython*, and that in your current working directory there is a subdirectory already created named **data**.

```python
f = open('data/romeo_juliet.csv', 'wb')

f.write('Word, Frequency\n')
for word, frequency in table:
    f.write('{w}, {f}\n'.format(w=word, f=frequency))
```

We first opened the file with the *open* command, but this time we passed it a second argument, *'wb'*, which roughly means *write data in the file*, instead of reading it. 

**Caution:** *python* won't ask for your permition before erasing the file if it already exists. It is assumed that *you know what you are doing*

After opening the file, we write a header line, and then use a cycle to write each entry in the frequency table in the format stablished by the header.
